```python
from tweepy import OAuthHandler
from tweepy import Stream
from tweepy.streaming import StreamListener
import time
```

```python
# setup security 
access_token = ''
access_token_secret = ''
consumer_key = ''
consumer_secret = ''

# setup file store
file = '/Users/Paulo/Documents/PUC/EXTRACAO INFORMACAO WEB/twitter/my-tweets-domingo-sample-9.json'
```

```python
class MyListener(StreamListener):
        
    
    def __init__(self, file):
        global the_file
        self.the_file = open(file, 'w')
        
    def on_data(self, data):
        self.the_file.write(data)
        return True
    
    def on_error(self, status):
        print(status)
        
    def file_close(self):
        self.the_file.close()        
```

```python
listener = MyListener(file)
oauth = OAuthHandler(consumer_key, consumer_secret)
oauth.set_access_token(access_token, access_token_secret)

max_time = 180 # in seconds
start_time = time.time() 

## collecting...
stream = Stream(oauth, listener)
stream.filter(track=['mbappe', 'neymar', 'messi','cr7'],async=True,languages=['pt'])
#stream.filter(track=['World Cup',"world cup","World cup","world Cup","FIFA World Cup","fifa world cup", "Fifa World Cup", "#WorldCup","#FIFA"],async=True)
#stream.filter(track=['FRANCEXCROATIA',"FRANCE X CROATIA","CROATIAXFRANCE","CROATIA X FRANCE","#CROATIA","#FRANCE", "CroatiaXFrance", "Croatia X France","FranceXCroatia","France X Croatia","#France","Croatia","FrancaXCroacia","CroaciaXFranca"],async=True)

elapsed_time = (time.time() - start_time)
while  elapsed_time < max_time:
    elapsed_time = (time.time() - start_time)

    
print('**Finish**')
stream.disconnect()
listener.file_close()
```

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tweets_data = []
tweets_file = open(file, "r")
for line in tweets_file:
    try:
        tweet = json.loads(line)
        tweets_data.append(tweet)
    except:
        continue
        
```

```python
tweets = pd.DataFrame()

tweets['lang'] = ''
tweets['text'] = ''
tweets['country'] = ''
for tweet in tweets_data:
    if 'lang' in tweet:
        tweets.loc[len(tweets)] = [tweet['lang'],
                               tweet['text'],
                               tweet['place']['country'] if tweet['place'] != None else None]
    
print(tweets.head(10))
```

```python
tweets_by_lang = tweets['country'].value_counts()

fig, ax = plt.subplots()
ax.tick_params(axis='x', labelsize=15)
ax.tick_params(axis='y', labelsize=10)
ax.set_xlabel('Languages', fontsize=15)
ax.set_ylabel('Number of tweets' , fontsize=15)
ax.set_title('Top 5 languages', fontsize=15, fontweight='bold')
tweets_by_lang[:5].plot(ax=ax, kind='bar', color='red')
```

```python
tweets_by_country = tweets['country'].value_counts()

fig, ax = plt.subplots()
ax.tick_params(axis='x', labelsize=15)
ax.tick_params(axis='y', labelsize=10)
ax.set_xlabel('Countries', fontsize=15)
ax.set_ylabel('Number of tweets' , fontsize=15)
ax.set_title('Top 5 countries', fontsize=15, fontweight='bold')
tweets_by_country[:5].plot(ax=ax, kind='bar', color='blue')
```

```python
import re
def word_in_text(word, text):
    word = word.lower()
    text = text.lower()
    match = re.search(word, text)
    if match:
        return True
    return False
```

```python
## filtrando aqueles q mais foram falados (termos mais influentes)
players = []
for tweet in tweets_data:
    if 'text' in tweet:
        if word_in_text('mbappe', tweet['text']):
            players.append('mbapper')
        elif word_in_text('neymar', tweet['text']):
            players.append('neymar')
        elif word_in_text('messi', tweet['text']):
            players.append('#messi')
        elif word_in_text('cr7', tweet['text']):
            players.append('cr7')
        else:
            players.append('None')
tweets['player'] = players
tweets.head(20)
```

```python
tweets_by_player = tweets['player'].value_counts()

fig, ax = plt.subplots()
ax.tick_params(axis='x', labelsize=15)
ax.tick_params(axis='y', labelsize=10)
ax.set_xlabel('Players', fontsize=15)
ax.set_ylabel('Number of tweets' , fontsize=15)
ax.set_title('Top 5 players', fontsize=15, fontweight='bold')
tweets_by_player[:6].plot(ax=ax, kind='bar', color='green')
```
